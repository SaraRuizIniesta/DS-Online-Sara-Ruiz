# Anlisis de Errores

Una parte que no debemos descuidar es el anlisis de errores y ms si conocemos el impacto de estos en nuestro problema de negocio. A continuacin vamos a ver de una forma introductoria como acometer esta tarea, plantendola para problemas de clasificacin primero y de regresin despus (ya que las mtricas y los objetivos son diferentes y por tanto la visualizacin de los errores tambin)

Para trabajar ambas situaciones usaremos el dataset `diamonds` que nos permite plantear problemas de ambos tipos, cargalo junto con las bibliotecas necesarias:

```python
import bootcampviztools as bt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor 


df = pd.DataFrame(sns.load_dataset("diamonds"))
```

Te dejo aqu una funcin (cortesa de ChatGPT y modificaciones de un servidor) para pintar las y_reales vs las y_predichas en regresin (lo usaremos en ese apartado)

```python
def plot_predictions_vs_actual(y_real, y_pred):
    """
    Funcin para graficar los valores reales vs. los valores predichos en una regresin.

    Args:
    y_real (array-like): Valores reales de la variable objetivo.
    y_pred (array-like): Valores predichos de la variable objetivo.
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(y_pred, y_real, alpha=0.5)
    plt.xlabel("Valores Predichos")
    plt.ylabel("Valores Reales")

    # Lnea y=x
    max_value = max(max(y_real), max(y_pred))
    min_value = min(min(y_real), min(y_pred))
    plt.plot([min_value, max_value], [min_value, max_value], 'r')

    plt.title("Comparacin de Valores Reales vs. Predichos")
    plt.show()

# Ejemplo de uso
# plot_predictions_vs_actual(y_real, y_pred)
```

### Errores en Clasificadores

Escogemos la variable target, que te sonars de otras pildoras y ejercicios:

```python
target_clasificacion = "cut"
```

```python
bt.pinta_distribucion_categoricas(df,[target_clasificacion], mostrar_valores= True, relativa= True)
```

Est desbalanceado pero no vamos a implementar tcnicas especiales porque para la ilustracin del anlisis nos sirve as.

#### Split de datos y tratamiento de features

```python
train_set, test_set = train_test_split(df, test_size= 0.2, random_state= 42)
```

Tratamos las variables categricas de clasificacin

```python
features_cat_clf = ["color","clarity"] 
```

En este caso tambin hay que convertir el target, lo vamos a hacer usando el mtodo `map` de pandas para que tengas otra forma de hacerlo (podramos haber usado el ordinal enconder de `sklearn`):

```python
mapeo_cut = {
    "Fair": 1,
    "Good": 2,
    "Very Good": 3,
    "Premium": 4,
    "Ideal": 5
}
train_set[target_clasificacion] = train_set[target_clasificacion].map(mapeo_cut)
test_set[target_clasificacion] = test_set[target_clasificacion].map(mapeo_cut)
```

```python
bt.pinta_distribucion_categoricas(train_set, [target_clasificacion], mostrar_valores= True, relativa= True)
```

Y ahora hacemos el onehot encoding empleando el get_dummies (quiz "clarity" estara mejor con un codificado ordinal pero, por no complicar la sesin, lo dejamos as)

```python
train_set = pd.get_dummies(train_set, columns= features_cat_clf, dtype = int)
test_set = pd.get_dummies(test_set, columns= features_cat_clf, dtype = int)
```

Tratamos las variables numricas para  clasificacin

```python
features_num_clf = [col for col in train_set.columns if col not in features_cat_clf and col != target_clasificacion]
```

```python
scaler_cat = StandardScaler()
train_set[features_num_clf] = scaler_cat.fit_transform(train_set[features_num_clf])
test_set[features_num_clf] = scaler_cat.transform(test_set[features_num_clf])
```

#### Modelado y Evaluacin

Y ahora ya podemos instanciar y entrenar nuestro modelo y ver qu tal es su comportamiento:

```python
X_train_clf = train_set.drop(target_clasificacion, axis = 1)
y_train_clf = train_set[target_clasificacion]
X_test_clf = test_set.drop(target_clasificacion, axis = 1)
y_test_clf = test_set[target_clasificacion]
```

```python
lr_clf = LogisticRegression(max_iter = 1000)
lr_clf.fit(X_train_clf, y_train_clf)
y_pred_clf = lr_clf.predict(X_test_clf)
print(classification_report(y_test_clf, y_pred_clf))
```

### Errores en Regresiones

#### Preparacin de datos

Hagamos algunas modificaciones sencillas para convertir el problema de clasificacin en uno de regresin. Basta con cambiar el target, que ahora ser el precio del diamante, y cambiar el juego de features predictoras:

```python
sns.histplot(x= "price", data = df, kde = True)
```

```python
target_regresion = "price"
features_cat_reg = features_cat_clf + [target_clasificacion]
features_num_reg = features_num_clf.copy()
features_num_reg.remove("price")

# Invertimos el escalado del precio
train_set["price"] = scaler_cat.inverse_transform(train_set[features_num_clf])[:,features_num_clf.index("price")]
test_set["price"] = scaler_cat.inverse_transform(test_set[features_num_clf])[:,features_num_clf.index("price")]
```

```python
sns.histplot(x= "price", data = train_set, kde = True)
```

#### Modelado y Evaluacin


Creamos los juegos X,y ahora que hemos cambiado el target y las features:

```python
X_train_reg = train_set.drop(target_regresion, axis = 1)
y_train_reg = train_set[target_regresion]
X_test_reg = test_set.drop(target_regresion, axis = 1)
y_test_reg = test_set[target_regresion]
```

```python
rf_reg = RandomForestRegressor(max_depth = 5, random_state= 42)
rf_reg.fit(X_train_reg, y_train_reg)
```

Evaluamos:

```python
y_pred_reg = rf_reg.predict(X_test_reg)
print("MAE:", mean_absolute_error(y_test_reg, y_pred_reg))
print("MAPE:", mean_absolute_percentage_error(y_test_reg, y_pred_reg))
print("RMSE:", np.sqrt(mean_squared_error(y_test_reg, y_pred_reg)))
```

#### Anlisis de Errores

En el caso de las regresiones no disponemos de una matriz de confusin pero s de los errores individuales como una variable continua a la que se le pueden analizar propiedades y hacer un par de visualizaciones:

**$y_{real}$ vs. $y_{predicha}$**

```python
plot_predictions_vs_actual(y_test_reg, y_pred_reg)
```

Podemos observar que el modelo tiene un comportamiento regular a precios "bajo" y a partir de los 7500 la dispersin de errores es ms grande y por franjas.

Cmo podemos tratarlo? Pues a posteriori o haciendo un modelo slo para precios superiores a 7500 y pasndolo por segunda vez a nuestra prediccin o bien avisando a Negocio y buscando soluciones... Lo que s es que tenemos esa informacin de cmo de bueno y malo es nuestro modelo actual (sin optimizar hiperparmetros)

**Residuos**

```python
# Calcular residuos
residuos = y_test_reg - y_pred_reg

# Visualizacin de residuos
plt.figure(figsize=(10, 6))
sns.histplot(residuos, kde=True)
plt.title('Distribucin de Residuos')
plt.xlabel('Error de Prediccin')
plt.ylabel('Frecuencia')
plt.show()
```

Son errores distribuidos normalmente, y eso suele ser "bueno" para las regresiones, pero es importante completar el anlisis viendo la grfica inicial de otra manera.

Podemos terminar de ver la dispersin de los residuos, si existen patrones el modelo est sesgando y deberamos tratarlo.

```python
# Grfica de dispersin de residuos
plt.scatter(y_pred_reg, residuos)
plt.title('Residuos vs. Predicciones')
plt.xlabel('Predicciones')
plt.ylabel('Residuos')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

Claramente deberamos pensar en qu hacer con esos puntos en los que la prediccin se concentra pero los valores reales son muy dispersos. Fjate que los problemas comienzan ms o menos cuando comienza la cola de la distribucin de precios. En este caso el anlisis de errores nos sugiere o hacer dos modelos uno para precios bajos y otros para altos, o bien binnear el precio y convertirla en un problema de clasificacin.
