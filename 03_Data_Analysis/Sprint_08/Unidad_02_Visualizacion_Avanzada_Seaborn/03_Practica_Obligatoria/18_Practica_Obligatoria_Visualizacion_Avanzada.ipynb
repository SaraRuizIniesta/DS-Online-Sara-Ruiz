Escojo las siguientes columnas:
longd(longitud) -> eje x
latd(latitud)-> eje y
city -> columna categorica para los colores

```python
def scatter_6(df, x_col, y_col, cat_col=None, size_or_col=60, escala=1.0, usar_colorbar=False):
   
    if isinstance(size_or_col, str):
        sizes = pd.to_numeric(df[size_or_col], errors="coerce") * escala
    else:
        sizes = size_or_col

    plt.figure(figsize=(10, 6))

    
    if cat_col is None:
        sns.scatterplot(data=df, x=x_col, y=y_col, s=sizes, legend=False)

    else:
        
        if usar_colorbar:
            codes = df[cat_col].astype("category").cat.codes
            sc = plt.scatter(df[x_col], df[y_col], c=codes, s=sizes, alpha=0.8)
            plt.colorbar(sc, label=cat_col)
        else:
            sns.scatterplot(data=df, x=x_col, y=y_col, hue=cat_col, s=sizes, legend=True)

    plt.title(f"{y_col} vs {x_col}")
    plt.tight_layout()
    plt.show()
```

```python
scatter_6(df, "longd", "latd", "city", 40, 1.0, usar_colorbar=True)
```

```python
def pie_chart(values, labels, title, filename=None, figsize=(5,5)):
    plt.figure(figsize=figsize)
    plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90)
    plt.title(title)
    plt.tight_layout()

    if filename is not None:
        plt.savefig(filename, dpi=200, bbox_inches="tight")

    plt.show()
```

```python
# 1. Mortalidad alta
surv_rate = df["survived"].mean()
vals = [1 - surv_rate, surv_rate]
labels = ["Died", "Survived"]

pie_chart(vals, labels, "Titanic: survival vs mortality")
```

```python
# Women & children firt: supervivencia por grupo

counts = df[df["survived"] == 1]["who"].value_counts()
pie_chart(counts.values, counts.index.tolist(), "Survivors distribution by who")
```

```python
counts = df[df["survived"] == 1]["pclass"].value_counts().sort_index()
pie_chart(counts.values, counts.index.astype(str).tolist(), "Survivors distribution by class (pclass)")
```

```python
# Diferencias tasa de superviviencia de mujeres y nios
df_wc = df[df["who"].isin(["child", "woman"])].copy()

for group in ["child", "woman"]:
    sub = df_wc[df_wc["who"] == group]["survived"].value_counts().sort_index()
    vals = [sub.get(0,0), sub.get(1,0)]
    labels = ["Died", "Survived"]
    pie_chart(vals, labels, f"Survival within group: {group}")
```

```python
df_emb = df.dropna(subset=["embarked"]).copy()

# 5a) Puertos entre supervivientes
counts = df_emb[df_emb["survived"] == 1]["embarked"].value_counts()
pie_chart(counts.values, counts.index.tolist(), "Survivors distribution by embarked")

# 5b) Puertos entre fallecidos
counts = df_emb[df_emb["survived"] == 0]["embarked"].value_counts()
pie_chart(counts.values, counts.index.tolist(), "Deaths distribution by embarked")
```

```python
for port in sorted(df_emb["embarked"].unique()):
    sub = df_emb[df_emb["embarked"] == port]["pclass"].value_counts().sort_index()
    pie_chart(sub.values, sub.index.astype(str).tolist(), f"Class distribution within embarked: {port}")
```

```python
import os
out_dir = "03_Practica_Obligatoria/figs"
os.makedirs(out_dir, exist_ok=True)
print("Carpeta creada:", out_dir)
```

```python
def pie_chart(values, labels, title, filename, figsize=(5,5)):
    plt.figure(figsize=figsize)
    plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(filename, dpi=200, bbox_inches="tight")
    plt.show()
```

```python
# (A) 1. Mortalidad alta
surv_rate = df["survived"].mean()
vals = [1 - surv_rate, surv_rate]
labels = ["Died", "Survived"]
pie_chart(vals, labels, "Titanic: survival vs mortality",
          f"{out_dir}/slide1_mortality.png")


# (B) 2. Women & children first (distribucin de supervivientes por who)
counts = df[df["survived"] == 1]["who"].value_counts()
pie_chart(counts.values, counts.index.tolist(), "Survivors distribution by who",
          f"{out_dir}/slide2_survivors_by_who.png")


# (C) 3. Clase influye (supervivientes por pclass)
counts = df[df["survived"] == 1]["pclass"].value_counts().sort_index()
pie_chart(counts.values, counts.index.astype(str).tolist(), "Survivors distribution by class (pclass)",
          f"{out_dir}/slide3_survivors_by_class.png")


# (D) 4. Diferencias mujeres vs nios (una tarta por grupo)
df_wc = df[df["who"].isin(["child", "woman"])].copy()

for group in ["child", "woman"]:
    sub = df_wc[df_wc["who"] == group]["survived"].value_counts().sort_index()
    vals = [sub.get(0,0), sub.get(1,0)]
    labels = ["Died", "Survived"]
    pie_chart(vals, labels, f"Survival within group: {group}",
              f"{out_dir}/slide4_survival_within_{group}.png")


# (E) 5. Embarked (pierdo las 2 filas sin puerto)
df_emb = df.dropna(subset=["embarked"]).copy()

counts = df_emb[df_emb["survived"] == 1]["embarked"].value_counts()
pie_chart(counts.values, counts.index.tolist(), "Survivors distribution by embarked",
          f"{out_dir}/slide5_survivors_by_embarked.png")

counts = df_emb[df_emb["survived"] == 0]["embarked"].value_counts()
pie_chart(counts.values, counts.index.tolist(), "Deaths distribution by embarked",
          f"{out_dir}/slide5_deaths_by_embarked.png")


# (E2 - opcional) Puerto como proxy de clase: una tarta por puerto
for port in sorted(df_emb["embarked"].unique()):
    sub = df_emb[df_emb["embarked"] == port]["pclass"].value_counts().sort_index()
    pie_chart(sub.values, sub.index.astype(str).tolist(), f"Class distribution within embarked: {port}",
              f"{out_dir}/slide5_class_within_embarked_{port}.png")
```
