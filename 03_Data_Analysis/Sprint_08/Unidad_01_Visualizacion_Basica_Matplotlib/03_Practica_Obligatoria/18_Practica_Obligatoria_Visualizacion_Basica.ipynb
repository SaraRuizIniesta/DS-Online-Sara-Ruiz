#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Grfico barras
top5 = df[['city', 'population_total']].dropna().sort_values('population_total', ascending=False).head(5)
ciudades = top5['city'].tolist()
habitantes = (top5['population_total'] / 1000).tolist()

# Hist rea total
area_total = df["area_total_km2"].dropna()

# Scatter agua vs poblacin
df_plot = df[["area_water_km2", "population_total"]].dropna()
x = df_plot["area_water_km2"]
y = df_plot["population_total"] / 1000
outliers = df_plot.nlargest(2, "area_water_km2")

# Boxplot altitud
elev = df["elevation_m"].dropna()

# --- Figura 2x2 ---
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# (1) Barras
colores = ["steelblue", "gray", "olive", "saddlebrown", "peru"]
bars = axs[0, 0].bar(ciudades, habitantes, color=colores, alpha=0.6)
axs[0, 0].set_title("Ciudades ms pobladas de California")
axs[0, 0].set_xlabel("Ciudades")
axs[0, 0].set_ylabel("Habitantes (miles)")
for bar, val in zip(bars, habitantes):
    axs[0, 0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 40, f"{val:.3f}", ha="center")

# (2) Histograma
axs[0, 1].hist(area_total, bins=50, alpha=0.5, color="steelblue", label="Hist Area")
axs[0, 1].set_title("Histograma del rea total (km)")
axs[0, 1].set_xlabel("Superficie (km)")
axs[0, 1].set_ylabel("Nmero de ciudades")
axs[0, 1].legend(loc="upper right")

# (3) Scatter
axs[1, 0].scatter(x, y, color="green", alpha=0.6)
axs[1, 0].set_title("Poblacin por kilmetro cuadrado de agua")
axs[1, 0].set_xlabel("rea de agua en la ciudad (km)")
axs[1, 0].set_ylabel("Poblacin (miles habitantes)")
axs[1, 0].grid(True, alpha=0.2)

# Flechas a outliers
axs[1, 0].text(0.60, 0.85, "Outliers", transform=axs[1, 0].transAxes)
for _, row in outliers.iterrows():
    ox = row["area_water_km2"]
    oy = row["population_total"] / 1000
    axs[1, 0].annotate(
        "",
        xy=(ox, oy),
        xycoords="data",
        xytext=(0.60, 0.85),
        textcoords="axes fraction",
        arrowprops=dict(arrowstyle="->", lw=1.2)
    )

# (4) Boxplot
axs[1, 1].boxplot(elev, vert=False, whis=1.5, patch_artist=True, boxprops=dict(facecolor="red"))
axs[1, 1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
axs[1, 1].set_xlabel("Altitud (m)")

fig.tight_layout()
```
